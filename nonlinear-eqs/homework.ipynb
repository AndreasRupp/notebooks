The figure shows the basin of attraction of $z=1$ when solving $z^3 = 1$ in the complex plane using Newton's method. The basin of attaction of $z=1$ is the set of points $z_0$ for which the method converges to $1$ when started with the initial guess $z_0$.

The figure is created as follows: for each point $z_0$ in the rectangle

$$
R = \{x + i y \in \mathbb C \mid x,y \in [-1,1] \}
